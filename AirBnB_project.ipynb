## 4. Data Cleaning

```python
dataset.isnull().sum()
```

```python
# drop all missing values
dataset.dropna(inplace=True)
```

```python
dataset.duplicated().sum()
```

```python
# drop all duplicated values
dataset.drop_duplicates(inplace=True)
```

```python
dataset.dtypes
```

```python
# changing data types
dataset['id'] = dataset['id'].astype(object)
dataset['host_id'] = dataset['host_id'].astype(object)
```

## 5. Data Analysis

#### A. Univariate Analysis

```python
# price Distribution

sns.histplot(data=dataset , x='price')
plt.show()
```

```python
# identifying Outliers

sns.boxplot(data=dataset , x='price')
plt.show()
```

```python
# removing Outliers

newData = dataset[dataset['price']<1500]
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=newData , x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=newData , x='availability_365')
plt.title('availability Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
newData.groupby(by='neighbourhood_group')['price'].mean()
```

```python
newData.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
newData['price per bed'] = newData['price'] / newData['beds']
```

```python
newData.head()
```

#### B. Multivariate Analysis

```python
sns.barplot(data=newData, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

```python
sns.scatterplot(data=newData, x='number_of_reviews', y='price',  hue='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data=newData, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10, 7))
sns.scatterplot(data=newData, x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = newData[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
plt.show()
```
